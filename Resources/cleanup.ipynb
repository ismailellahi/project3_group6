```python
# Import dependencies
import pandas as pd
from pathlib import Path

# Name of the CSV file
file = Path('basic_incidents2021.csv')
```

```python
# Read csv file
df= pd.read_csv(file)
df.head()
```

```python
# drop rows without coridnates( long and latitude not given)
newdf = df[df['geometry'] != "{'type': 'Point', 'coordinates': (0.0, 0.0)}"]
newdf.head(2)
```

```python
# remove other characters except coordinates
newdf2 = newdf.replace({'geometry' : "{'type': 'Point', 'coordinates':"}, {'geometry' : ""}, regex = True).replace({'geometry' : "}"}, {'geometry' : ""}, regex = True)
newdf2['geometry'] = newdf2['geometry'].str.replace('(', '')
newdf2['geometry'] = newdf2['geometry'].str.replace(')', '')
```

```python
# drop unwanted column
df3 = newdf2.drop(["Ward_At_Event_Dispatch"], axis = 1)
```

```python
# Split columns with ID and description
df3[["Final_Incident_ID", "Final_Incident_Types"]] = df3['Final_Incident_Type'].str.split('-', 1, expand = True)

df3[["Call_ID", "Call_Origin"]] = df3['Call_Source'].str.split('-', 1, expand = True)

df3[["Location", "Location2"]] = df3['Intersection'].str.split('/', 1, expand = True)

df3.head(2)
```

```python
# List all the column names
df3.columns
```

```python
# Rearrange the columns for new data frame
df_fire = df3[['_id', 'INCIDENT_NUMBER', 'Initial_CAD_Event_Type',
       'Initial_CAD_Event_Call_Type', 'Final_Incident_ID', 'Final_Incident_Types',
       'Event_Alarm_Level', 'Call_ID', 'Call_Origin','Incident_Station_Area',
       'Incident_Ward', 'Location', 'TFS_Alarm_Time', 'TFS_Arrival_Time',
       'Last_TFS_Unit_Clear_Time', 'Persons_Rescued', 'geometry']]
       
df_fire.head(2)
       
```

```python
# drop rows with empty values
df_fire2= df_fire.dropna(axis = 0)
```

```python
# change format of columns with date/time
from datetime import datetime
df_fire2["TFS_Alarm_Time"] = pd.to_datetime(df_fire2["TFS_Alarm_Time"])
df_fire2["TFS_Arrival_Time"] = pd.to_datetime(df_fire2["TFS_Arrival_Time"])
df_fire2["Last_TFS_Unit_Clear_Time"] = pd.to_datetime(df_fire2["Last_TFS_Unit_Clear_Time"])
```

```python
# find the time taken and time at scene
df_fire2["Time_Taken_to_Arrive"] = df_fire2["TFS_Arrival_Time"] - df_fire2["TFS_Alarm_Time"]
df_fire2["Total_Time_at_Scene"]= df_fire2["Last_TFS_Unit_Clear_Time"] - df_fire2["TFS_Arrival_Time"]
```

```python
# convert total time in minutes
df_fire2["Time_Taken_to_Arrive"] = df_fire2["Time_Taken_to_Arrive"].dt.seconds/60
df_fire2["Total_Time_at_Scene"] = df_fire2["Total_Time_at_Scene"].dt.seconds/60
```

```python
# check info 
df_fire2.info()
```

```python
df_fire2.head()
```

```python
# rearrange columns in the final data frame
df_final1 = df_fire2[['_id', 'INCIDENT_NUMBER', 'Initial_CAD_Event_Type',
       'Initial_CAD_Event_Call_Type', 'Final_Incident_ID', 'Final_Incident_Types',
       'Event_Alarm_Level', 'Call_ID', 'Call_Origin','Incident_Station_Area',
       'Incident_Ward', 'Location', 'TFS_Alarm_Time', 'TFS_Arrival_Time', 'Time_Taken_to_Arrive', 
       'Last_TFS_Unit_Clear_Time', 'Total_Time_at_Scene', 'Persons_Rescued', 'geometry']]

df_final1.head(3)
```

```python
# Split longitude and latitude into separate columns
df_final1[["Longitude", "Latitude"]] = df_fire2['geometry'].str.split(',', 1, expand = True)
```

```python
# change data type for longitude and latitude 
df_final1["Latitude"] = df_final1["Latitude"].astype(float)
df_final1["Longitude"] = df_final1["Longitude"].astype(float)
```

```python
df_final =df_final1[['_id', 'INCIDENT_NUMBER', 'Initial_CAD_Event_Type',
       'Initial_CAD_Event_Call_Type', 'Final_Incident_ID', 'Final_Incident_Types',
       'Event_Alarm_Level', 'Call_ID', 'Call_Origin','Incident_Station_Area',
       'Incident_Ward', 'Location', 'TFS_Alarm_Time', 'TFS_Arrival_Time', 'Time_Taken_to_Arrive', 
       'Last_TFS_Unit_Clear_Time', 'Total_Time_at_Scene', 'Persons_Rescued', 'Latitude', 'Longitude']]

df_final.head(3)
```

```python
# change data type for other columns
df_final["_id"] = df_final["_id"].astype(int)
df_final1["Latitude"] = df_final1["Latitude"].astype(float)
df_final["Time_Taken_to_Arrive"]= df_final['Time_Taken_to_Arrive'].astype(str)
df_final["Total_Time_at_Scene"]= df_final['Total_Time_at_Scene'].astype(str)

df_final.info()
```

```python
# extract month from the date
df_final["Month"] = pd.to_datetime(df_final["TFS_Alarm_Time"]).dt.strftime('%m')
df_final["Month"]= df_final["Month"].astype(int)
```

```python
df_finally = df_final.replace({"Total_Time_at_Scene": "0 days 00:"}, {'Total_Time_at_Scene' : ""}, regex = True)
df_finally1 = df_finally.replace({"Total_Time_at_Scene": ":"}, {'Total_Time_at_Scene' : "."}, regex = True)

df_finally2 = df_finally1.replace({"Time_Taken_to_Arrive": "0 days 00:"}, {'Time_Taken_to_Arrive' : ""}, regex = True)

df_finally3 = df_finally2.replace({"Time_Taken_to_Arrive": ":"}, {'Time_Taken_to_Arrive' : "."}, regex = True)
# df_final2 = df_final1.replace({"Time_Taken_to_Arrive": "0 days 00:"}, {'Time_Taken_to_Arrive' : ""}, regex = True)

```

```python
# change data type in new data columns
df_finally3["Time_Taken_to_Arrive"]= round(df_finally3["Time_Taken_to_Arrive"].astype(float, 3))
df_finally3["Total_Time_at_Scene"]= round(df_finally3["Total_Time_at_Scene"].astype(float, 3))
df_finally3.head(3)
```

```python
# save new data frame as a csv file
df_finally3.to_csv("dataset.csv", encoding='utf8', index=False)

```
